```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
#line plot
line1=range(10)
line2=[x*x for x in line1]
plt.plot(line1,line2)
plt.show()
```

```python
line1=np.arange(1,20,2)
line2=np.arange(1,20,2)
plt.title("line plot")
plt.xlabel("range 1-20 in steps of 2")
plt.ylabel("range 1-20 in steps of 2")
plt.plot(line1,line2)
plt.savefig("G:\Extra SEMESTER\plot1.jpg")
plt.show()
```

```python
#sine graph
line1=np.arange(0,10,0.1)
line2=np.sin(line1)
print(line2)
plt.title("sine curve")
plt.xlabel("0 to 10 in steps of 0.1")
plt.ylabel("sin value")
plt.plot(line1,line2)
plt.savefig("G:\Extra SEMESTER\sine.jpg")
plt.show()
```

```python
line1=np.arange(1,10,0.1)
line2=np.sin(line1)
line3=np.cos(line1)
#print(line2)
plt.title("sin curve")
plt.xlabel("0 to 10 in steps of 0.1")
plt.ylabel("-1 to 1")
plt.plot(line1,line2,line1,line3)
#adding legends
plt.legend(['sin','cos'])
plt.savefig("G:\Extra SEMESTER\plot3.jpg")
plt.show()
```

```python
#scatter chart
x=np.arange(0,100,20)
y=np.arange(50,-50,-20)
y1=np.arange(100,0,-20)
plt.title("with labels different colors")
plt.xlabel("range 0- 100")
plt.ylabel("range 50- -50")
plt.scatter(x,y,label="one",color="green")
plt.scatter(x,y1,label="two",color="red)
plt.legend(['a','b'])
plt.savefig("G:\Extra SEMESTER\a.jpg")
plt.show()
```

```python
#create lsits
x=np.arange(0,100,20)
y=np.arange(50,-50,-20)
y1=np.arange(100,0,-20)
plt.figure(figsize=(5,5))
plt.title("with grid lines and sized")
plt.xlabel("range 0 to 100")
plt.ylabel("range 50 to -50")
plt.xticks(np.linspace(0,100,10,endpoint=True))
plt.grid(True)
plt.scatter(x,y,label="one",color="green")
plt.scatter(x,y1,label="two",color="red")
plt.legend(['a','b'])
plt.show()
```

```python
#twin axes plot
figure=plt.figure()
axes1=figure.add_subplot(1,1,1)
axes2=axes1.twinx()
line1=np.linspace(0.,10.,100)
line2=[x**2 for x in line1]
line3=[x**3 for x in line1]
axes1.plot(line1,line2,'b-')
axes2.plot(line1,line3,'r-')
axes1.set_ylabel('x power 2',color="red")
axes2.set_ylabel('X*x',color="blue")
axes1.set_xlabel('X')
plt.show()
```
